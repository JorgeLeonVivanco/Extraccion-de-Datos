```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

*Filtros basicos para variables cualitativas*

```python
# 1.- Filtro por cliente específico
filtro_cliente = df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro_cliente.to_csv("filtro_cliente.csv")
print(filtro_cliente.head())
```

```python
# 2.- Filtro por filas específicas

# Variación 1: Filtrar filas de la 3 a la 10
filtro_filas_var1 = df.iloc[3:11, :]
filtro_filas_var1.to_csv("filtro_filas_var1.csv", index=False)
print(filtro_filas_var1)

# Variación 2: Filtrar filas 1, 4 y 7
filtro_filas_var2 = df.iloc[[1, 4, 7], :]
filtro_filas_var2.to_csv("filtro_filas_var2.csv", index=False)
print(filtro_filas_var2)
```

```python
# 3.- Filtros por columnas
# filtro_columnas1: Muestra la columna 1 y 3 pero todas las filas
filtro_columnas1 = df.iloc[:, 1:3]
filtro_columnas1.to_csv("filtro_columnas1.csv", index=False)
print(filtro_columnas1)

# filtro_columnas2: Muestra las columnas 1, 4 y 5 pero todas las filas
filtro_columnas2 = df.iloc[:, [1, 4, 5]]
filtro_columnas2.to_csv("filtro_columnas2.csv", index=False)
print(filtro_columnas2)
```

```python
# Leer el archivo con una columna específica como índice
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=1)  # Si no se tiene un índice, Python genera uno automáticamente
print(df1.head())  # Mostrar las primeras filas con el nuevo índice
```

```python
# 4.-Filas vs columnas: Filtrar por clientes específicos y una columna relevante
filtro_filas_columnas = df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]["PRECIO_UNITARIO"]
filtro_filas_columnas.to_csv("filtro_filas_columnas.csv", index=False)
print(filtro_filas_columnas)
```

```python
# 5.-Filtro general: Mostrar las primeras filas del DataFrame
filtro_general = df.head()
filtro_general.to_csv("filtro_general.csv", index=False)
print(filtro_general)
```

```python
# 6.-Filtro de las últimas filas del DataFrame
filtro_final = df.tail()
filtro_final.to_csv("filtro_final.csv", index=False)
print(filtro_final)
```

*Filtros basicos para variables cuantitativas*

```python
# 7.-Filtros de comparación: Filtrar registros donde PRECIO_UNITARIO sea mayor a 7000
filtro_comparacion = df[df["PRECIO_UNITARIO"] > 10000]
filtro_comparacion.to_csv("filtro_comparacion.csv")
print(filtro_comparacion)
```

*Filtros basicos con conectores logicos*

```python
# 8.-Filtros con condiciones múltiples (AND)
filtro_y = df[(df["PRECIO_UNITARIO"] < 6000) & (df["FECHA_DOC"] == "2022-01-05")]
filtro_y.to_csv("filtro_y.csv", index=False)
print(filtro_y)
```

```python
# 9.-Filtros con condiciones múltiples (OR)
filtro_o = df[(df["PRECIO_UNITARIO"] < 6000) | (df["FECHA_DOC"] == "2023-03-14")]
filtro_o.to_csv("filtro_o.csv", index=False)
print(filtro_o)
```

```python
# 10.-Filtros de exclusión: Excluir registros con una fecha específica
filtro_exclusion = df[~(df["FECHA_DOC"] == "2023-03-14")]
filtro_exclusion.to_csv("filtro_exclusion.csv", index=False)
print(filtro_exclusion)
```
